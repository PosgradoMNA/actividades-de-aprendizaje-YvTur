##### **Erasing missing data**

```python
#Pandas has a method that drops the row or the column that has a missing value
# Axis=0 is a row
# Axis=1 is a column
#by adding the inplace, we make the change in the dataset
df.dropna(subset=["price"], axis=0, inplace= True)
df
```

##### **Replacing missing data**

```python
#In this specific case, some data dissapear, whick generate that the data type was object
#due the dataframe is a small one I manually replace it,
#The technique we are using here is for replacing the values in case there is a missing one
#I was unable to use this technique because we had question mark,
#so, the .mean funciton was unable to work
mean = df["normalized-losses"].mean()
df["normalized-losses"].replace(np.nan, mean)
```

##### **Incorrect data types**

```python
#We are going to review the column price, which should be int64
df["price"].tail(5)
```

```python
#We need to review with more detail the types of data. 
#In the above case we see an object, while should we expect an int
df["price"] = df["price"].astype(int)
```

##### **Data Normalization**

At moment of making comparisson is complicated to have things in different ranges of 0s. 
Meaning you can have something in the range of 1 to 10 and another column in the range from 10 to 100. 

Normalization is having everything in the same range for making it easier to do proper calculations

There are different normalization:
*   Simple feature scalling
*   Min-Max
*   Z-score






```python
#lets apply this for the lenght feature
df["length"] = df['length']/df["length"].max()
df
```

#### **Binning**

Es generar caterogiras dentro de un mismo rango. En este caso podemos designar la columna de precio. Se podra indicar categorias como barato, medio y caro. *texto en cursiva*

```python
from numpy.core.function_base import linspace
#COde to make bings in a column
bins=np.linspace(min(df["price"]),max(df["price"]),4)
group_names=["Low","Medium","High"]
df["price-binned"]=pd.cut(df["price"],bins,labels=group_names,include_lowest=True)

df
```
